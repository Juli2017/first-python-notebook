Analysis of California Prop 67 from November 2016 election

```python
import pandas as pd
```

Reading in the list of propositions

```python
props = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
```

```python
props.head()
```

```python
props.head()
```

```python
props.info()
```

Reading in the list of contributions for the ballot measures

```python
contribs = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

```python
contribs.contributor_lastname = contribs.contributor_lastname.fillna('Missing_lastname')
```

```python
contribs.contributor_firstname = contribs.contributor_firstname.fillna('Missing_firstname')
```

```python
contribs.head()
```

```python
contribs.info()
```

```python
props.prop_name.value_counts().reset_index()
```

```python
props.ccdc_committee_id.value_counts().reset_index()
```

```python
props.prop_name.value_counts()
```

```python
prop_67 = props[props.prop_name == 'PROPOSITION 067- REFERENDUM TO OVERTURN BAN ON SINGLE-USE PLASTIC BAGS.']
```

```python
prop_67.head()
```

```python
prop_67.info()
```

```python
merged = pd.merge(prop_67, contribs, on='calaccess_committee_id')
```

```python
merged.head()
```

```python
merged.info()
```

```python
merged.amount.sum()
```

```python
merged.committee_position.value_counts()
```

```python
support = merged[merged.committee_position == "SUPPORT"]
```

```python
oppose = merged[merged.committee_position == "OPPOSE"]
```

```python
support.amount.sum()
```

```python
oppose.amount.sum()
```

```python
percent_support = support.amount.sum() / (support.amount.sum() + oppose.amount.sum())
```

```python
percent_support
```

```python
merged.sort_values("amount", ascending=False)
```

```python
oppose.sort_values("amount", ascending=False).head(10)
```

```python
merged.groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
merged.groupby(["contributor_state", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```
